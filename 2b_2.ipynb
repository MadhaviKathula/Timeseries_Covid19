```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load your dataset
vaccination_data = pd.read_csv('us_state_vaccinations.csv')
```

```python
# Convert 'date' to datetime type
vaccination_data['date'] = pd.to_datetime(vaccination_data['date'])
```

```python

# Calculate the national daily average of total vaccinations per hundred
national_daily_avg = vaccination_data.groupby('date')['total_vaccinations_per_hundred'].mean()
```

```python

# Calculate the daily changes and identify significant deviations
daily_changes = national_daily_avg.diff()
change_std = daily_changes.std()
significant_deviation_threshold = 2 * change_std
```

```python
# Identify significant deviations
significant_deviations = daily_changes[(daily_changes.abs() > significant_deviation_threshold)]
```

```python
# Plotting
plt.figure(figsize=(14, 7))
plt.plot(national_daily_avg.index, national_daily_avg, label='Daily Average of Total Vaccinations per Hundred', color='blue')
plt.scatter(significant_deviations.index, national_daily_avg[significant_deviations.index], color='red', label='Significant Deviations')
plt.title('National Average of Total Vaccinations per Hundred Over Time with Significant Deviations')
plt.xlabel('Date')
plt.ylabel('Total Vaccinations per Hundred')
plt.legend()
plt.grid(True)
plt.show()
```
